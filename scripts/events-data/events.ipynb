```python
# read this json file
path = "./eventbrite-huge.json"

import json

with open(path) as f:
    data = json.load(f)
```

```python
# remove crawled_at from each event
for event in data:
    del event['crawled_at']
```

```python
from datetime import datetime
def save_data(data):
    for event in data:
        # get current index
        index = data.index(event)
        print("index", index)
        
        name = event['name']

        import re
        file_friendly_name = re.sub(r'[^\w\s-]', '', name).strip().replace(' ', '_')

        with open(f"./events/{file_friendly_name}.txt", 'w') as file:

# Event Name: WET SATURDAYS
# Description: Wet Saturdays at Wet Lounge – The Hottest Hip-Hop and Open Format Party in Town!
# Date and Time: Starts on May 11, 2024, at 9:00 PM and ends on May 12, 2024, at 1:30 AM (Pacific Time).
# Location: WET LOUNGE, 18250 Hesperian Boulevard, San Lorenzo, CA 94580, USA
# Event Type: Social Event, Offline

            # break the loop if startDate is not present
            if 'startDate' not in event:
                print("Start Date not present")
                continue


            start_date = datetime.strptime(event['startDate'], "%Y-%m-%dT%H:%M:%S%z")
            end_date = datetime.strptime(event['endDate'], "%Y-%m-%dT%H:%M:%S%z")

            start_date_str = start_date.strftime("%B %d, %Y, at %I:%M %p")
            end_date_str = end_date.strftime("%B %d, %Y, at %I:%M %p")

            # get image if it exists
            image = 'No Image'
            if 'image' in event:
                image = event['image']
            
            # check if orgaziner exists
            organizer = 'No Organizer'
            if 'organizer' in event:
                if 'name' in event['organizer']:
                    organizer = event['organizer']['name']

            file.write(f"Event Name: {event['name']}\n")
            file.write(f"Description: {event['description']}\n")
            file.write(f"Date and Time: Starts on {start_date_str} and ends on {end_date_str} (Pacific Time).\n")
            file.write(f"Location: {event['location']['address']['streetAddress']}\n")
            file.write(f"URL: {event['url']}\n")
            file.write(f"Image: {image}\n")
            file.write(f"Organizer: {organizer}\n")


```

```python
save_data(data)
```
